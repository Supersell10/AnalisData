# 2. Паспорт датасета

```python
# Импорт библиотек

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from faker import Faker
import random

plt.style.use('ggplot')
sns.set(font_scale=1)

fake = Faker()

np.random.seed(42)
random.seed(42)
```

## Генерация датасета

```python
brands_models = {
    'Toyota': ['Corolla', 'Camry', 'RAV4'],
    'BMW': ['X5', '320', '520'],
    'Mercedes': ['C200', 'E200', 'GLA'],
    'Audi': ['A4', 'A6', 'Q5'],
    'Volkswagen': ['Golf', 'Passat', 'Tiguan'],
    'Ford': ['Focus', 'Mondeo', 'Explorer'],
    'Kia': ['Rio', 'Sportage', 'Cerato'],
    'Hyundai': ['Solaris', 'Tucson', 'Elantra']
}

fuel_types = ['Petrol', 'Diesel', 'Hybrid', 'Electric']

transmissions = ['Manual', 'Automatic']

colors = ['Black', 'White', 'Silver', 'Blue', 'Red', 'Gray']

rows = []

for i in range(1000):

    brand = random.choice(list(brands_models.keys()))

    model = random.choice(brands_models[brand])

    year = random.randint(2005, 2024)

    fuel = random.choice(fuel_types)

    transmission = random.choice(transmissions)

    mileage = max(0, int(np.random.normal(120000, 50000)))

    engine_volume = round(np.random.uniform(1.0, 5.0), 1)

    horsepower = int(engine_volume * np.random.uniform(70, 120))

    base_price = (
        300000
        + horsepower * 4000
        + (year - 2005) * 25000
    )

    price = int(np.random.normal(base_price, 150000))

    price = max(price, 200000)

    color = random.choice(colors)

    owners = random.randint(1, 5)

    registration_date = fake.date_between(
        start_date='-15y',
        end_date='today'
    )

    rows.append([
        i + 1,
        brand,
        model,
        year,
        fuel,
        transmission,
        mileage,
        engine_volume,
        horsepower,
        price,
        color,
        owners,
        registration_date
    ])

columns = [
    'car_id',
    'brand',
    'model',
    'year',
    'fuel_type',
    'transmission',
    'mileage',
    'engine_volume',
    'horsepower',
    'price',
    'color',
    'owners',
    'registration_date'
]

df = pd.DataFrame(rows, columns=columns)
```
